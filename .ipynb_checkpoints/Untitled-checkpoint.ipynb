$$F_{12} = G\frac{m_1m_2}{\left|r\right|^2}\hat{r}$$

$$F_{12} = G\frac{m_1m_2}{\left|r\right|^3}r$$

$$F_{12} = m_1a_1$$
$$F_{21} = m_2a_2$$

$$F_{21} = -F_{12}$$

$$\ddot{r_1} = G\frac{m_2}{\left|r_2-r_1\right|^3}(r_2-r_1)$$

$$\ddot{r_2} = -G\frac{m_1}{\left|r_2-r_1\right|^3}(r_2-r_1)$$

```julia
import LinearAlgebra: norm
import Plots: plot3d, plot3d!

G = 10
m1 = m2 = 10
r1 = [-5,0]
r2 = [5,0]
v1 = [0,-2]
v2 = [0,2]

h = .1
n = 500

rr1 = zeros(2,n)
rr2 = zeros(2,n)

rr1[:,1] = r1
rr2[:,1] = r2

for i=2:n
    r = r2 - r1
    r_norm = norm(r)
    F = G*m1*m2*r/r_norm^3
    a1 = F / m1
    a2 = -F / m2

    # update velocity
    v1 += h*a1
    v2 += h*a2

    # update position
    r1 += h*v1
    r2 += h*v2

    rr1[:,i] = r1
    rr2[:,i] = r2
end
```

```julia
plot(rr1[1,:], rr1[2,:])
plot!(rr2[1,:], rr2[2,:],color=2)
```

```julia
tt = [h*i for i=1:n]
plot3d(tt, rr1[1,:], rr1[2,:], ylim=(-15,15), zlim=(-15,15))
plot3d!(tt, rr2[1,:], rr2[2,:],color=2)
```

```julia
import LinearAlgebra: norm

G = 10
m1 = m2 = m3 = 10
r1 = [-5,2]
r2 = [0,0]
r3 = [5,-2]
v1 = [2,2]
v2 = [-4,-4]
v3 = [2,2]

h = .1
n = 100

rr1 = zeros(2,n)
rr2 = zeros(2,n)
rr3 = zeros(2,n)

rr1[:,1] = r1
rr2[:,1] = r2
rr3[:,1] = r3

for i=2:n
    r12 = r2 - r1
    r13 = r3 - r1
    r23 = r3 - r2

    r12_norm = norm(r12)
    r13_norm = norm(r13)
    r23_norm = norm(r23)

    F12 = G*m1*m2/r12_norm^3 * r12
    F13 = G*m1*m3/r13_norm^3 * r13
    F23 = G*m2*m3/r23_norm^3 * r23

    a1 = (F12 + F13) / m1
    a2 = (-F12 + F23) / m2
    a3 = (-F13 - F23) / m3

    # update velocity
    v1 += h*a1
    v2 += h*a2
    v3 += h*a3

    # update position
    r1 += h*v1
    r2 += h*v2
    r3 += h*v3

    # update solution matrix
    rr1[:,i] = r1
    rr2[:,i] = r2
    rr3[:,i] = r3
end
```

```julia
plot(rr1[1,:], rr1[2,:])
plot!(rr2[1,:], rr2[2,:],color=2)
plot!(rr3[1,:], rr3[2,:],color=3)
```

```julia
tt = [h*i for i=1:n]
plot3d(tt, rr1[1,:], rr1[2,:], ylim=(-15,15), zlim=(-15,15))
plot3d!(tt, rr2[1,:], rr2[2,:],color=2)
plot3d!(tt, rr3[1,:], rr3[2,:],color=3)
```
